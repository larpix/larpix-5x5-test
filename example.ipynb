```python
import numpy
import h5py
import matplotlib.pyplot as plt
```

```python
# open data file
f = h5py.File('data/packet-2025_09_10_15_12_07_PDT.h5')

# select 'packets' dataset
p = f['packets']

# select data packets (packet_type = 1)
d = p[p['packet_type'] == 1]

```

```python
# plot dataword distribution per channel

fig, axes = plt.subplots(8, 8, figsize=(17, 12), sharex=True)  # Create a figure and a grid of subplots

# loop over the 64 channels
for channel_id in range(64): 

    # select only entries corresponding to channel_id
    d_channel = d[d['channel_id'] == channel_id]

    # plot a histogram of the pedestal datawords 
    axes[channel_id//8, channel_id%8].hist(d_channel['dataword'], range=[250, 300], linewidth=0.5, bins=50, histtype='step')
    
    axes[channel_id//8, channel_id%8].set_title(f'Channel {channel_id}')

fig.supxlabel('dataword [ADC Count]')
fig.supylabel('counts')
plt.tight_layout()
```
